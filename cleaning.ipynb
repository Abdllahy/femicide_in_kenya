```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# import openpyxl as px
```

```python
# import csv
df = pd.read_csv('femicide-data_2016_2023.csv')
```

```python
# make a copy of the dataframe
df_cleaned = df.copy()
```

```python
df_cleaned
```

```python
# renaming columns
# df.rename(columns = {

# })
```

```python
# drop empty columns
df_cleaned = df.drop(columns=['Unnamed: 24', 'Unnamed: 25'])
```

```python
df_cleaned
```

```python
# drop unnecessary columns
df_cleaned = df_cleaned.drop(columns=['url', 'title', 'text'])
```

```python
df_cleaned = df_cleaned.drop(columns=['country_name'])
```

```python
df_cleaned
```

```python
# replace placeholder values with NaN
placeholders = ['Unknown', 'unknown', "", 'N/A', 'n/a']
df_cleaned.replace(placeholders, np.nan, inplace=True)
```

```python
# replace whitespace with NaN
df_cleaned = df_cleaned.replace(r'^\s*$', np.nan, regex=True)
```

```python
df_cleaned
```

```python
# drop empty rows
# what is the number of empty rows?
df_cleaned.isna().all(axis=1).sum()
```

```python
# drop all the empty rows
df_cleaned = df_cleaned.dropna(how='all')
```

```python
# check the dataframe
df_cleaned
```

```python
# so how many columns do we have now?
print(df_cleaned.shape[1])
# i have 20 columns including the index
```

```python
# checking the data types of these columns
df_cleaned.dtypes
```

```python
# we can see our data types are all objects meaning they are all in strings

# let's change the data types of the columns to their appopriate data types
```

```python
# but before that, let's standardize the column names
df_cleaned.columns = (
    df_cleaned.columns
    .str.strip() # remove leading and trailing whitespace
    .str.lower() # convert to lowercase
    .str.replace(' ', '_') # replace spaces with underscores
)
```

```python
df_cleaned
```

```python
# let's now proceed to change the data types of the columns
# create a list of the date columns
date_col = [
    'published_date', 
    'date_of_murder', 
    'verdict_date', 
    'court_date_(first_appearance)'
]

for col in date_col:
    df_cleaned.loc[:, col] = pd.to_datetime(df_cleaned[col], errors='coerce', dayfirst=True).dt.date

```

```python
df_cleaned._is_copy # checks if df_cleaned is a copy of the original data
```

```python
# let's now convert the other columns to numeric
numeric_cols = ["age", "years_of_sentence", "days_to_verdict"]

# Loop through and convert
for col in numeric_cols:
    if col == "days_to_verdict":
        # Remove commas before converting
        df_cleaned.loc[:, col] = pd.to_numeric(df_cleaned[col].astype(str).str.replace(",", ""), errors="coerce")
    else:
        df_cleaned.loc[:, col] = pd.to_numeric(df_cleaned[col], errors="coerce")
```

```python
df_cleaned.dtypes[["age", "years_of_sentence", "days_to_verdict"]]
```

```python
print("AGE:", df_cleaned["age"].unique()[:20])
print("YEARS OF SENTENCE:", df_cleaned["years_of_sentence"].unique()[:20])
print("DAYS TO VERDICT:", df_cleaned["days_to_verdict"].unique()[:20])
```

```python
for col in ["age", "years_of_sentence", "days_to_verdict"]:
    df_cleaned[col] = df_cleaned[col].astype(float)
```

```python
df_cleaned.dtypes[["age", "years_of_sentence", "days_to_verdict"]]
```

```python
df_cleaned.dtypes
```

```python
# split location column into location and county
split_location = df_cleaned['location'].str.split(',', n=1, expand=True)

# Assign using .loc to avoid the warning
df_cleaned.loc[:, 'area'] = split_location[0].str.strip()
df_cleaned.loc[:, 'county'] = split_location[1].str.strip()
# remove leading and trailing whitespace
```

```python
df_cleaned
```

```python
# i am seeing location and area, let's overwrite location with area
df_cleaned.drop(columns = ['location'], inplace=True)
```

```python
df_cleaned
```

```python
df_cleaned.dtypes
```

```python
# # let's remove the time in the date columns
for col in date_col:
    df_cleaned[col] = pd.to_datetime(df_cleaned[col], errors='coerce', dayfirst=True).dt.date
```

```python
# save the cleaned data to a new csv file
df_cleaned.to_csv('new_data.csv', index=False)
```
